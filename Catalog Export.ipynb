```python
import re

filepath = "C:\Users\Jamie\Untitled Folder\catalog_bottom.txt"
f = open(filepath)
data = f.readlines()
f.close()

print(data[:10])
```

```python
list = []
num_title_lines = 0

for each in data:
    x = re.match("([A-Z][A-Z][A-Z]\s[A-Z][0-9][0-9])", each)
    if x:
        list.append(each)
        num_title_lines = 1
    elif num_title_lines > 0 & num_title_lines < 4:
        list.append(each)
        num_title_lines+=1
    if num_title_lines == 4:
        num_title_lines = 0

sorted_list = [list[x:x+4] for x in range(0, len(list),4)]

str_list = []

for each in sorted_list:
    str = ""
    for item in each:
        item = item.rstrip()
        str += item + " "
    str_list.append(str)
        
str_list[:20]
        
```

```python
final_str = ""

for each in str_list:
    final_str += each + " \n"
    
text_file = open("booklist.txt", "w")
text_file.write(final_str)
text_file.close()
```
